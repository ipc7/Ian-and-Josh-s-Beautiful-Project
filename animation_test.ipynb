```python
from __future__ import division
import numpy as np
from numpy import sin,cos,pi
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.path import Path
import matplotlib.patches as patches

def generalSystem2d(S,t,G=6.67408e-11):
    #S includes m formatted mi,xi,yi,vxi,vyi
    #G=6.67408e-11
    #print(S0)
    N=len(S)//5
    x = S[1::5]
    y = S[2::5]
    vx = S[3::5]
    vy = S[4::5]
    m = S[0::5]    

    xr3 = np.zeros((N,N))
    yr3 = np.zeros((N,N))

    dvx = np.zeros(N)
    dvy = np.zeros(N)

    dm = np.zeros(N)
            
    for i in range(N):
        for j in range(N):
            if(i!=j):
                r = np.sqrt((x[i]-x[j])**2+(y[i]-y[j])**2)
                xr3[i][j] = (x[i]-x[j])/(r**3)
                yr3[i][j] = (y[i]-y[j])/(r**3)
                
                dvx[i] = dvx[i]-G*m[j]*xr3[i][j]
                dvy[i] = dvy[i]-G*m[j]*yr3[i][j]


    return_array = np.array([])

    for i in range(N):
        return_array = np.append(return_array,[dm[i],vx[i],vy[i],dvx[i],dvy[i]])

    return return_array
```
